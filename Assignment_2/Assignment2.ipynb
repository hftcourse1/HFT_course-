```python
import pandas as pd
import numpy as np
import os
import statsmodels.api as sm
import matplotlib.pyplot as plt
save_path = fr"...\HFT_course\Assignment_2\plots\\"
```

```python
def analysis(df, tar_log = "log_osv", target = 'avg_osv', tar = "<osv>", 
             save_path=save_path, name="img"):
    """
    This function does the entire analysis for the power-law estimates.
    """
    
    x = df["log_thrs"]
    y = df[tar_log]
    
    X2 = sm.add_constant(x)
    est = sm.OLS(y, X2)
    results = est.fit()
    print(results.summary())
    
    a = results.params[0]
    b = results.params[1]
    
    E_os = b
    C = 1/(np.exp(a/b))

    print()
    print("E =", E_os, "+/-", round((results.conf_int().iloc[1].diff()/2)[1], 5))
    print("E*log(1/C) =", a)
    print("log(1/C) =", a/b)
    print("C =", C)
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))
    # generate x-values for your regression line (two is sufficient)
    min_x = np.floor(np.min(df.log_thrs)) + 0.001
    max_x = np.ceil(np.max(df.log_thrs))+0.2
    x = np.linspace(min_x, max_x, num=150)
    # scatter-plot data
    df.plot(x="log_thrs", y=tar_log, kind='scatter', ax=ax[0])
    # plot regression line on the same axes, set x-axis limits
    ax[0].plot(x, a + b * x, color="r")
    ax[0].set_xlim([min_x - 0.00001, max_x])
    ax[0].set_title('Linear fit on log-data', fontsize = 16)
    ax[0].set_xlabel("log(Δx_dc)")
    ax[0].set_ylabel(f"log({tar})")
    ax[0].grid()
    
    min_x = np.floor(np.min(df.thrs)) + 0.001
    max_x = np.max(df.thrs) + 0.001
    x = np.linspace(min_x, max_x, num=150)
    # scatter-plot data
    df.plot(x="thrs", y=target, kind='scatter', ax=ax[1])
    # plot regression line on the same axes, set x-axis limits
    ax[1].plot(x, (x/C)**E_os, color="r")
    ax[1].set_xlim([min_x - 0.001, max_x])
    ax[1].set_title('Power-law fit on data', fontsize = 16)
    ax[1].set_xlabel("Δx_dc")
    ax[1].set_ylabel(f"{tar}")
    ax[1].grid()
    
    plt.tight_layout()

    if not(name is None):
        plt.savefig(save_path + name)
    plt.show()
```

```python
data_path = fr"...\HFT_course\Data\Assignment2\BTC"
data = os.listdir(data_path)

df_up = pd.DataFrame(columns=["thrs", "avg_osv", "N"])
df_down = pd.DataFrame(columns=["thrs", "avg_osv", "N"])

for i in data:
    if "Threshold_down" in i:
        path = data_path + fr"/{i}"
        df=pd.read_csv(path, sep=",", names=["thrs", "osv"])
        
        df_min = np.min(df.index[df["thrs"] == 10000.0])
        idx_drop = df.index[df_min + 1:]
        
        df = df.drop(idx_drop)

        if not(all(i[1] == 10000 for i in df.iloc[-1].items())):
            raise ValueError(f"There is a checksum for {df.thrs[0]} that does not add up")
        else:
            df.drop(index=df.index[-1], axis=0, inplace=True)

        n = float(len(df))
        # print(path, df.osv.mean())
        df_down = pd.concat((df_down, pd.DataFrame.from_dict({"thrs":[df.thrs[0]], "avg_osv":[df.osv.mean()], "N":n})))
    elif "Threshold_up" in i:
        path = data_path + fr"/{i}"
        df=pd.read_csv(path, sep=",", names=["thrs", "osv"])

        df_min = np.min(df.index[df["thrs"] == 10000.0])
        idx_drop = df.index[df_min + 1:]
        
        df = df.drop(idx_drop)
        
        if not(all(i[1] == 10000 for i in df.iloc[-1].items())):
            raise ValueError(f"There is a checksum for {df.thrs[0]} that does not add up")
        else:
            df.drop(index=df.index[-1],axis=0,inplace=True)
        
        n = float(len(df))
        df_up = pd.concat((df_up, pd.DataFrame.from_dict({"thrs":[df.thrs[0]], "avg_osv":[df.osv.mean()], "N":n})))
    else:
        raise ValueError("Problems")


df_down = df_down.set_index("thrs", drop = False)
df_up = df_up.set_index("thrs", drop = False)

df_all = pd.concat([df_down, df_up], join="outer", axis=1)
df_all.avg_osv = df_all.avg_osv.mean(axis=1)
df_all.N = df_all.N.mean(axis=1)
df_all = df_all.loc[:, ~df_all.columns.duplicated()]
```

### For up-ward trend:

```python
analysis(df_up, name="up_trend")
```

### For down-ward trend:

```python
analysis(df_down, name="down_trend")
```

```python
analysis(df_all, name="all_trend")
```
