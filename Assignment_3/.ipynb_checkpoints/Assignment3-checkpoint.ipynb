```python
import pandas as pd
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt
```

```python
def loader(path):
    df=pd.read_csv(path, sep=",", names=["thrs", "osv"])
    df_min = np.min(df.index[df["thrs"] == 10000.0])
    idx_drop = df.index[df_min: -1]
    df = df.drop(idx_drop)
    if not(all(i[1] == 10000 for i in df.iloc[-1].items())):
        raise ValueError(f"There is a checksum for {df.thrs[0]} that does not add up")
    else:
        df.drop(index=df.index[-1], axis=0, inplace=True)
    return df
```

```python
path = fr"...\HFT_course\Data\Assignment2\BTC\Threshold_down_0.001.txt"
df_short = loader(path)
path = fr"...\HFT_course\Data\Assignment2\BTC\Threshold_up_0.001.txt"
df_long = loader(path)
```

```python
win_fast = 25
win_slow = 250
th = 0.001
a = 1; b = 1


def bayes(win, th):
    return lambda x: ((x > th).sum() + a)/(win + a + b)

df_short[f"rol_ex_{win_fast}"] = df_short.osv.rolling(window=win_fast).apply(bayes(win_fast, th), engine="cython")
df_short[f"rol_ex_{win_slow}"] = df_short.osv.rolling(window=win_slow).apply(bayes(win_slow, th), engine="cython")
df_short["diff_rol"] = df_short[f"rol_ex_{win_fast}"] - df_short[f"rol_ex_{win_slow}"]

df_long[f"rol_ex_{win_fast}"] = df_long.osv.rolling(window=win_fast).apply(bayes(win_fast, th), engine="cython")
df_long[f"rol_ex_{win_slow}"] = df_long.osv.rolling(window=win_slow).apply(bayes(win_slow, th), engine="cython")
df_long["diff_rol"] = df_long[f"rol_ex_{win_fast}"] - df_long[f"rol_ex_{win_slow}"]
```

```python
save_path = fr"...\HFT_course\Assignment_3\plots\prop_plot.png"

fig, ax = plt.subplots(1, 2, figsize=(16, 8))


ax[0].set_title("Winning probability for short trading", fontsize=16)
ax[0].plot(df_short[f"rol_ex_{win_fast}"])
ax[0].plot(df_short[f"rol_ex_{win_slow}"])
# ax[0].legend()
ax[0].grid()
ax[0].set_ylim([0, 1])

ax[1].set_title("Winning probability for long trading", fontsize=16)
ax[1].plot(df_long[f"rol_ex_{win_fast}"])
ax[1].plot(df_long[f"rol_ex_{win_slow}"])
# ax[1].legend()
ax[1].grid()
ax[1].set_ylim([0, 1])

fig.supylabel("Winning probability: p")
fig.supxlabel("Event number")

    
    
plt.tight_layout()
plt.savefig(save_path)
plt.show()
```
